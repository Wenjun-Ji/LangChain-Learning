```python
# sparkllm
os.environ["IFLYTEK_SPARK_APP_ID"] = ""
os.environ["IFLYTEK_SPARK_API_KEY"] = ""
os.environ["IFLYTEK_SPARK_API_SECRET"] = ""
#　此处参考：https://www.xfyun.cn/doc/spark/Web.html
os.environ["IFLYTEK_SPARK_API_URL"] = ""
os.environ["IFLYTEK_SPARK_llm_DOMAIN"] = ""

from langchain_community.chat_models import ChatSparkLLM

model = ChatSparkLLM()
```

[HumanMessage](https://api.python.langchain.com/en/latest/output_parsers/langchain_core.output_parsers.string.StrOutputParser.html)

[SystemMessage](https://api.python.langchain.com/en/latest/messages/langchain_core.messages.system.SystemMessage.html)

```python
from langchain_core.messages import HumanMessage, SystemMessage

messages = [
    SystemMessage(content="请把下面的内容翻译为英文"),
    HumanMessage(content="我是你爹"),
]

model.invoke(messages)
```

## [OutputParsers](https://api.python.langchain.com/en/latest/output_parsers/langchain_core.output_parsers.string.StrOutputParser.html)

```python
from langchain_core.output_parsers import StrOutputParser

parser = StrOutputParser()
```

StrOutputParser()：使用它的一种方法是单独使用它。例如，我们可以保存语言模型调用的结果，然后将其传递给解析器。
model.invoke()返回值包括 字符串响应和其他元数据的 "AIMessage" 对象
StrOutputParser的输出是一个字符串,可以把result的其他元数据过滤掉，只返回响应字符串

```python
result = model.invoke(messages)
print(result) 
parser.invoke(result) 
```

更常见的是，我们可以用这个输出解析器“链接”模型。这意味着此输出解析器将在此链中每次被调用。该链采用语言模型的输入类型（字符串或消息列表），并返回输出解析器的输出类型（字符串）。
我们可以使用运算符轻松创建链。在LangChain中，运算符用于将两个元素组合在一起。| |

```python
chain = model | parser
chain.invoke(messages)
```
