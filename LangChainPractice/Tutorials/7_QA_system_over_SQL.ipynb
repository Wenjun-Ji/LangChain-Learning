# 基于 SQL 数据构建问答系统

使大型语言模型（LLM）系统查询结构化数据与查询非结构化文本数据有本质上的不同。对于后者，通常会生成可以在向量数据库中搜索的文本，而对于结构化数据，通常是让LLM编写并执行一种DSL（如SQL）的查询。在本指南中，我们将介绍如何在数据库中的表格数据上创建一个问答系统。我们将覆盖使用链和代理的实现。这些系统将允许我们对数据库中的数据提出问题并获得自然语言的答案。两者之间的主要区别在于，我们的代理可以在需要时循环查询数据库多次以回答问题。

构建SQL数据库的问答系统需要执行模型生成的SQL查询。这有一定的风险。确保您的数据库连接权限始终根据链/代理的需要进行最小范围的限制。这将减轻但不能完全消除构建模型驱动系统的风险。有关更多的一般安全最佳实践，请参阅这里。


### 架构
在高层次上，这些系统的步骤如下：

1. **将问题转换为DSL查询**：模型将用户输入转换为SQL查询。
2. **执行SQL查询**：执行查询。
3. **回答问题**：模型使用查询结果回应用户输入。

请注意，查询CSV数据可以采用类似的方法。有关如何在CSV数据上进行问答的详细信息，请参阅我们的指南。
